```python
import matplotlib.pyplot as plt
```

> # 1) line graph 

```python
x = [1,2,3,4]
y = [10,15,14,10]
plt.plot(x,y)
plt.show()
```

> # 2) scatter plot 

```python
x = [1,2,3,4]
y = [10,15,20,10]
plt.scatter(x,y)
plt.show()
```

```python
Name = ["bmw","honda" ,"audi"]
speed = [200,100,300]
plt.scatter(Name ,speed)
plt.show()
```

> ## modify the axis name 
    plt.xlabel("_____")
    plt.ylabel("_____")

```python
Name = ["bmw","honda" ,"audi"]
speed = [200,100,300]
plt.scatter(Name ,speed)
plt.xlabel("Car Name")
plt.ylabel("Speed of the car")
plt.show()
```

> # change the colour 
## plt.scatter(Name ,speed, __c= "red"__ / __color = "red__)

```python
Name = ["bmw","honda" ,"audi"]
speed = [200,100,300]
plt.scatter(Name ,speed,c= "red")
plt.xlabel("Car Name")
plt.ylabel("Speed of the car")
plt.show()
```

```python
Name = ["bmw","honda" ,"audi"]
speed = [200,100,300]
plt.plot(Name ,speed,c= "red")
plt.xlabel("Car Name")
plt.ylabel("Speed of the car")
plt.show()
```

> # linewidth  -- increase the size of the line present on the graph
###  plt.plot(Name ,speed,c= "red" , __linewidth = 3__)


```python
Name = ["bmw","honda" ,"audi"]
speed = [200,100,300]
plt.plot(Name ,speed,c= "red")
plt.xlabel("Car Name")
plt.ylabel("Speed of the car")
plt.show()
```

```python
Name = ["bmw","honda" ,"audi"]
speed = [200,100,300]
plt.plot(Name ,speed,c= "red" , linewidth = 3)
plt.xlabel("Car Name")
plt.ylabel("Speed of the car")
plt.show()
```

> # linestyle -- change the style of the line 
### plt.plot(Name ,speed,c= "red" , __linestyle = "--"__)

```python
Name = ["bmw","honda" ,"audi"]
speed = [200,100,300]
plt.plot(Name ,speed,c= "red" , linewidth = 3 , linestyle = "--")
plt.xlabel("Car Name")
plt.ylabel("Speed of the car")
plt.show()
```

> # Marker -- gives the data point on the grpah 
## plt.plot(Name ,speed,c= "red" , linewidth = 3 , linestyle = "--",__marker = "o"__)

```python
Name = ["bmw","honda" ,"audi"]
speed = [200,100,300]
plt.plot(Name ,speed,c= "red" , linewidth = 3 , linestyle = "--",marker = "o")
plt.xlabel("Car Name")
plt.ylabel("Speed of the car")
plt.show()
```

> # markersize -- size of the marker will be increased 
### plt.plot(Name ,speed,c= "red" , linewidth = 3 , linestyle = "--",marker = "o" , __markersize = 10__) 


```python

Name = ["bmw","honda" ,"audi"]
speed = [200,100,300]
plt.plot(Name ,speed,c= "red" , linewidth = 3 , linestyle = "--",marker = "o" , markersize = 10)
plt.xlabel("Car Name")
plt.ylabel("Speed of the car")
plt.show()
```

> # msc -- marker edge colur 
    outside colour of the marker
### plt.plot(Name ,speed,c= "red" , linewidth = 3 , linestyle = "--",marker = "o" , markersize = 10 ,__mec = "green"__)

```python

Name = ["bmw","honda" ,"audi"]
speed = [200,100,300]
plt.plot(Name ,speed,c= "red" , linewidth = 3 , linestyle = "--",marker = "o" , markersize = 10 ,mec = "green")
plt.xlabel("Car Name")
plt.ylabel("Speed of the car")
plt.show()
```

> # mfc -- > marker face colour 
    front colour  of the marker
### plt.plot(Name ,speed,c= "red" , linewidth = 3 , linestyle = "--",marker = "o" , markersize = 10 ,mec = "green" , __mfc = "yellow"__)


```python

Name = ["bmw","honda" ,"audi"]
speed = [200,100,300]
plt.plot(Name ,speed,c= "red" , linewidth = 3 , linestyle = "--",marker = "o" , markersize = 10 ,mec = "green" , mfc = "yellow")
plt.xlabel("Car Name")
plt.ylabel("Speed of the car")
plt.show()
```

```python

Name = ["bmw","honda" ,"audi"]
speed = [200,100,300]
plt.plot(Name ,speed,c= "red" , linewidth = 3 , linestyle = "--",marker = "*" , markersize = 10 ,mec = "green" , mfc = "yellow")
plt.xlabel("Car Name")
plt.ylabel("Speed of the car")
plt.show()
```

> # title 
## __plt.title(_ __ )__

```python

Name = ["bmw","honda" ,"audi"]
speed = [200,100,300]
plt.plot(Name ,speed,c= "red" , linewidth = 3 , linestyle = "--",marker = "*" , markersize = 10 ,mec = "green" , mfc = "yellow")
plt.title("car and their speed ")
plt.xlabel("Car Name")
plt.ylabel("Speed of the car")
plt.show()
```

> # grid 
## __plt.grid()__

```python

Name = ["bmw","honda" ,"audi"]
speed = [200,100,300]
plt.plot(Name ,speed,c= "red" , linewidth = 3 , linestyle = "--",marker = "*" , markersize = 10 ,mec = "green" , mfc = "yellow")
plt.title("car and their speed ")
plt.xlabel("Car Name")
plt.ylabel("Speed of the car")
plt.grid()
plt.show()
```
